````python
from pathlib import Path
from shutil import rmtree
from typing import Union, List, Dict
from tqdm import tqdm

import requests
from llama_cpp import Llama, llama_tokenizer
````

–ó–∞–≥—Ä—É–∑–∫–∞ –≤–µ—Å–æ–≤ –º–æ–¥–µ–ª–∏ –≤ —Ñ–æ—Ä–º–∞—Ç–µ GGUF - –≤–∞—Ä–∏–∞–Ω—Ç —á–µ—Ä–µ–∑ `requests`

````python
# —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –∑–∞–≥—Ä—É–∑–∫–∏ —Ñ–∞–π–ª–∞ –ø–æ URL —Å—Å—ã–ª–∫–µ –∏ –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ–º –ø—Ä–æ–≥—Ä–µ—Å—Å –±–∞—Ä–∞
def download_file(file_url: str, file_path: Union[str, Path]) -> None:
    response = requests.get(file_url, stream=True)
    if response.status_code != 200:
        raise Exception(f'–§–∞–π–ª –Ω–µ–¥–æ—Å—Ç—É–ø–µ–Ω –¥–ª—è —Å–∫–∞—á–∏–≤–∞–Ω–∏—è –ø–æ —Å—Å—ã–ª–∫–µ: {file_url}')
    # –ø–æ–ª—É—á–µ–Ω–∏–µ —Ä–∞–∑–º–µ—Ä–∞ —Ñ–∞–π–ª–∞ –≤ –±–∞–π—Ç–∞—Ö
    total_size = int(response.headers.get('content-length', 0))
    # –ø—Ä–æ–≥—Ä–µ—Å—Å –±–∞—Ä tqdm
    progress_bar = tqdm(desc='–ó–∞–≥—Ä—É–∑–∫–∞ —Ñ–∞–π–ª–∞ GGUF', total=total_size, unit='iB', unit_scale=True)
    # –æ—Ç–∫—Ä—ã—Ç–∏–µ —Ñ–∞–π–ª–∞ –Ω–∞ –∑–∞–ø–∏—Å—å
    with open(file_path, 'wb') as file:
        # –∑–∞–≥—Ä—É–∑–∫–∞ —Ñ–∞–π–ª–∞ —á–∞–Ω–∫–∞–º–∏ –ø–æ 4–ö–± –∏ –∑–∞–ø–∏—Å—å —á–∞–Ω–∫–∞ –≤ —Ñ–∞–π–ª
        for data in response.iter_content(chunk_size=4096):
            size = file.write(data)
            # –æ–±–Ω–æ–≤–ª–µ–Ω–∏–µ –ø—Ä–æ–≥—Ä–µ—Å—Å –±–∞—Ä–∞
            progress_bar.update(size)

# –ø—Ä—è–º–∞—è —Å—Å—ã–ª–∫–∞ –Ω–∞ –º–æ–¥–µ–ª—å –≤ —Ñ–æ—Ä–º–∞—Ç–µ GGUF
# gguf_url = 'https://huggingface.co/bartowski/Qwen2.5-1.5B-Instruct-GGUF/resolve/main/Qwen2.5-1.5B-Instruct-Q8_0.gguf'
# gguf_url = 'https://huggingface.co/bartowski/openchat-3.6-8b-20240522-GGUF/resolve/main/openchat-3.6-8b-20240522-IQ4_XS.gguf'
# gguf_url = https://huggingface.co/bartowski/gemma-2-2b-it-GGUF/resolve/main/gemma-2-2b-it-IQ3_M.gguf
# gguf_url = 'https://huggingface.co/bartowski/Llama-3.2-1B-Instruct-GGUF/resolve/main/Llama-3.2-1B-Instruct-Q8_0.gguf'
gguf_url = 'https://huggingface.co/bartowski/gemma-2-2b-it-GGUF/resolve/main/gemma-2-2b-it-Q8_0.gguf'

# –ø–æ–ª–Ω—ã–π –ø—É—Ç—å –¥–æ –º–æ–¥–µ–ª–∏ –≤ –ø–∞–ø–∫–µ ./models
model_path = Path('models') / gguf_url.rsplit('/')[-1]
model_path.parent.mkdir(exist_ok=True)

# –µ—Å–ª–∏ –º–æ–¥–µ–ª—å –æ—Ç—Å—É—Ç—Å—Ç–≤—É–µ—Ç –≤ –ø–∞–ø–∫–µ models —Ç–æ –Ω—É–∂–Ω–æ –µ–µ –∑–∞–≥—Ä—É–∑–∏—Ç—å
if not model_path.is_file():
    if not gguf_url.endswith('.gguf'):
        raise Exception('–°—Å—ã–ª–∫–∞ –Ω–∞ –º–æ–¥–µ–ª—å LLM –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –ø—Ä—è–º–æ–π —Å—Å—ã–ª–∫–æ–π –Ω–∞ —Ñ–∞–π–ª GGUF')
    else:
        print('–ó–∞–≥—Ä—É–∑–∫–∞ —Ñ–∞–π–ª–∞ GGUF –¥–ª—è LLM –º–æ–¥–µ–ª–∏ ...')
        download_file(gguf_url, model_path)
````

````python
from pathlib import Path
from shutil import rmtree
from typing import Union, List, Dict, Tuple, Optional
from tqdm import tqdm

import requests
import gradio as gr
from llama_cpp import Llama


# CHAT_HISTORY = list[list[str | GradioComponent | tuple[str] | tuple[str | Path, str] | None]] | Callable | None
# CHAT_HISTORY = List[Tuple[Optional[Union[Tuple[str], str]], Optional[str]]]
CHAT_HISTORY = List[Tuple[Optional[str], Optional[str]]]
MODEL_DICT = Dict[str, Llama]


# =============== –§–£–ù–ö–¶–ò–ò –ó–ê–ì–†–£–ó–ö–ò –ú–û–î–ï–õ–ò ===============================

# —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –∑–∞–≥—Ä—É–∑–∫–∏ —Ñ–∞–π–ª–∞ –ø–æ URL —Å—Å—ã–ª–∫–µ –∏ –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ–º –ø—Ä–æ–≥—Ä–µ—Å—Å –±–∞—Ä–æ–≤ tqdm –∏ gradio
def download_file(file_url: str, file_path: Union[str, Path]) -> None:
    response = requests.get(file_url, stream=True)
    if response.status_code != 200:
        raise Exception(f'–§–∞–π–ª –Ω–µ–¥–æ—Å—Ç—É–ø–µ–Ω –¥–ª—è —Å–∫–∞—á–∏–≤–∞–Ω–∏—è –ø–æ —Å—Å—ã–ª–∫–µ: {file_url}')
    total_size = int(response.headers.get('content-length', 0))
    progress_tqdm = tqdm(desc='–ó–∞–≥—Ä—É–∑–∫–∞ —Ñ–∞–π–ª–∞ GGUF', total=total_size, unit='iB', unit_scale=True)
    progress_gradio = gr.Progress()
    completed_size = 0
    with open(file_path, 'wb') as file:
        for data in response.iter_content(chunk_size=4096):
            size = file.write(data)
            progress_tqdm.update(size)
            completed_size += size
            desc = f'–ó–∞–≥—Ä—É–∑–∫–∞ —Ñ–∞–π–ª–∞ GGUF, {completed_size/1024**3:.3f}/{total_size/1024**3:.3f} GB'
            progress_gradio(completed_size/total_size, desc=desc)


# —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –∑–∞–≥—Ä—É–∑–∫–∏ —Ñ–∞–π–ª–∞ –º–æ–¥–µ–ª–∏ –≤ —Ñ–æ—Ä–º–∞—Ç–µ GGUF –∏ –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏–∏ –Ω–æ–≤–æ–π –º–æ–¥–µ–ª–∏ –∏–∑ –Ω–µ–≥–æ
# —Ç–∞–∫ –∂–µ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –ª–æ–≥–∏ –∑–∞–≥—Ä—É–∑–∫–∏ –º–æ–¥–µ–ª–∏ –¥–ª—è –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è –≤ —ç–ª–µ–º–µ–Ω—Ç–µ gr.Textbox()
# –∏ —Ñ–ª–∞–≥ –ø–æ–¥–¥–µ—Ä–∂–∫–∏ —Å–∏—Å—Ç–µ–º–Ω–æ–≥–æ –ø—Ä–æ–º—Ç–∞ –º–æ–¥–µ–ª—å—é
def download_gguf_and_init_model(gguf_url: str, model_dict: MODEL_DICT) -> Tuple[MODEL_DICT, bool, str]:
    # —Å—Ç—Ä–æ–∫–∞ —Å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–µ–π –æ –∑–∞–≥—Ä—É–∑–∫–µ –∏ –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏–∏ –º–æ–¥–µ–ª–∏ –¥–ª—è –¥–∞–ª—å–Ω–µ–π—à–µ–≥–æ –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è –≤ gr.Textbox()
    log = ''
    # –ø—Ä–æ–≤–µ—Ä–∫–∞ —á—Ç–æ —Å—Å—ã–ª–∫–∞ –Ω–∞ GGUF –º–æ–¥–µ–ª—å –æ–∫–∞–Ω–∏—á–∏–≤–∞–µ—Ç—Å—è –Ω–∞ .gguf
    if not gguf_url.endswith('.gguf'):
        log += f'–°—Å—ã–ª–∫–∞ –Ω–∞ –º–æ–¥–µ–ª—å LLM –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –ø—Ä—è–º–æ–π —Å—Å—ã–ª–∫–æ–π –Ω–∞ —Ñ–∞–π–ª GGUF\n'
        return model_dict, log

    # –∏–∑–≤–µ—á–µ–Ω–∏–µ –∏–º–µ–Ω–∏ —Ñ–∞–π–ª–∞ –º–æ–¥–µ–ª–∏ –∏ —Ñ–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–∏–µ –ø—É—Ç–∏ –¥–æ –Ω–µ–µ
    gguf_filename = gguf_url.rsplit('/')[-1]
    model_path = MODELS_PATH / gguf_filename
    # –ø—Ä–æ–≥—Ä–µ—Å—Å –±–∞—Ä Gradio
    progress = gr.Progress()

    # –µ—Å–ª–∏ –º–æ–¥–µ–ª—å –µ—â–µ –Ω–µ –∑–∞–≥—Ä—É–∂–µ–Ω–∞ –≤ –ø–∞–ø–∫—É MODELS_PATH (./models)
    if not model_path.is_file():
        # –∑–∞–≥—Ä—É–∑–∫–∞ –º–æ–¥–µ–ª–∏ –ø–æ URL —Å—Å—ã–ª–∫–µ
        progress(0.3, desc='–®–∞–≥ 1/2: –ó–∞–≥—Ä—É–∑–∫–∞ –º–æ–¥–µ–ª–∏ GGUF')
        try:
            download_file(gguf_url, model_path)
            log += f'–§–∞–π–ª –º–æ–¥–µ–ª–∏ {gguf_filename} –∑–∞–≥—Ä—É–∂–µ–Ω\n'
        # –µ—Å–ª–∏ –∑–∞–≥—Ä—É–∑–∫–∞ –Ω–µ —É–¥–∞–ª–∞—Å—å –¥–æ –∑–∞–ø–∏—Å–∞—Ç—å –≤ –ª–æ–≥–∏ –∫–æ–¥ –æ—à–∏–±–∫–∏ –∏ –∏–Ω—Ñ–æ –æ —Ç–µ–∫—É—â–µ–π –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä–æ–≤–∞–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏
        except Exception as ex:
            log += f'–û—à–∏–±–∫–∞ –∑–∞–≥—Ä—É–∑–∫–∏ –º–æ–¥–µ–ª–∏ –ø–æ —Å—Å—ã–ª–∫–µ {gguf_url}, –∫–æ–¥ –æ—à–∏–±–∫–∏:\n{ex}\n'
            curr_model = model_dict.get('model')
            # –µ—Å–ª–∏ –ø–æ –∫–∞–∫–∏–º —Ç–æ –ø—Ä–∏—á–∏–Ω–∞–º –º–æ–¥–µ–ª—å –æ—Ç—Å—É—Å—Ç–≤—É–µ—Ç –≤ —Å–ª–æ–≤–∞—Ä–µ model_dict —Ç–æ –¥–æ–±–∞–≤–∏—Ç—å —ç—Ç–æ –≤ –ª–æ–≥–∏
            if curr_model is None:
                log += f'–ù–∏–∫–∞–∫–∞—è –º–æ–¥–µ–ª—å –Ω–µ –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä–æ–≤–∞–Ω–∞ –≤ —Å–ª–æ–≤–∞—Ä–µ model_dict\n'
                return model_dict, load_log
            # –Ω–∞–∑–≤–≤–∞–Ω–∏–µ —Ñ–∞–π–ª–∞ —Ç–µ–∫—É—â–µ–π –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ä–æ–≤–∞–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏
            curr_model_filename = Path(curr_model.model_path).name
            log += f'–¢–µ–∫—É—â–∞—è –∏–Ω–∏—Ü–∏–∞–ª–∏—â–∏—Ä–æ–≤–∞–Ω–Ω–∞—è –º–æ–¥–µ–ª—å: {curr_model_filename}\n'
            return model_dict, log
    # –µ—Å–ª–∏ –º–æ–¥–µ–ª—å —É–∂–µ –∑–∞–≥—Ä—É–∂–µ–Ω–∞ —Ç–æ –¥–æ–±–∞–≤–∏—Ç—å —ç—Ç–æ –≤ –ª–æ–≥–∏
    else:
        log += f'–§–∞–π–ª –º–æ–¥–µ–ª–∏ {gguf_filename} –∑–∞–≥—Ä—É–∂–µ–Ω, –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –º–æ–¥–µ–ª–∏...\n'

    # —É–¥–∞–ª–µ–Ω–∏–µ —Ñ–∞–π–ª–∞ –ø—Ä–µ–¥—ã–¥—É—â–µ–π –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä–æ–≤–∞–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏ –µ—Å–ª–∏ —Ç—Ä–µ–±—É–µ—Ç—Å—è —Ç–∞–∫–æ–π —Ñ—É–Ω–∫—Ü–∏–æ–Ω–∞–ª
    # curr_model = model_dict.get('model')
    # if curr_model is not None and Path(curr_model.model_path).is_file():
    #     Path(curr_model.model_path).unlink(missing_ok=True)

    # –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –∑–∞–≥—Ä—É–∂–µ–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏
    progress(0.7, desc='–®–∞–≥ 2/2: –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –º–æ–¥–µ–ª–∏')
    model = Llama(model_path=str(model_path), n_gpu_layers=-1, verbose=False)
    model_dict = {'model': model}
    # –ø–æ–¥–¥–µ—Ä–∂–∏–≤–∞–µ—Ç –ª–∏ –º–æ–¥–µ–ª—å —Å–∏—Å—Ç–µ–º–Ω—ã–π –ø—Ä–æ–º—Ç
    support_system_role = 'System role not supported' not in model.metadata['tokenizer.chat_template']
    log += f'–ú–æ–¥–µ–ª—å {gguf_filename} –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∏—Ä–æ–≤–∞–Ω–∞\n'
    return model_dict, support_system_role, log


# =============== –§–£–ù–ö–¶–ò–ò –ò–ù–¢–ï–†–§–ï–ô–°–ê –ß–ê–¢-–ë–û–¢–ê =============================

# –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ —Å–æ–æ–±—â–µ–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –≤ –æ–∫–æ—à–∫–æ —á–∞—Ç –±–æ—Ç–∞
def user_message_to_chatbot(user_message: str, chatbot: CHAT_HISTORY) -> Tuple[str, CHAT_HISTORY]:
    if user_message:
        chatbot.append((user_message, None))
    return '', chatbot


# –≥–µ–Ω–µ—Ä–∞—Ü–∏—è –æ—Ç–≤–µ—Ç–∞ –º–æ–¥–µ–ª—å—é –∏ –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –æ—Ç–≤–µ—Ç–∞ –≤ –æ–∫–Ω–æ —á–∞—Ç –±–æ—Ç–∞ –ø–æ—Å–∏–º–≤–æ–ª—å–Ω–æ
# —Ñ—É–Ω–∫—Ü–∏—è –±—É–¥–µ—Ç –≤—ã–∑—ã–≤–∞—Ç—Å—è –ø—Ä–∏ –Ω–∞–∂–∞—Ç–∏–∏ –Ω–∞ Enter –≤ –æ–∫–æ—à–∫–µ –≤–æ–≤–æ–¥–∞ —Ç–µ–∫—Å—Ç–∞
# –∫—Ä–æ–º–µ –æ–±—ã—á–Ω—ã—Ö –∞—Ä–≥—É–º–µ–Ω—Ç–æ–≤ - —Å–æ–æ–±—â–µ–Ω–∏—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –∏ –∏—Å—Ç–æ—Ä–∏–∏ - –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –µ—â–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è –∫–æ–Ω—Ñ–∏–≥–∞ –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏
def bot_response_to_chatbot(
        chatbot: CHAT_HISTORY,  # —Å–ø–∏—Å–æ–∫ –ø–µ—Ä–µ–ø–∏—Å–æ–∫ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –∏ –±–æ—Ç–∞, –∫–æ—Ç–æ—Ä—ã–π –æ—Ç–æ–±—Ä–∞–∂–∞–µ—Ç—Å—è –≤ –æ–∫–æ—à–∫–µ —á–∞—Ç-–±–æ—Ç–∞
        model_dict: MODEL_DICT,  # —Å–ª–æ–≤–∞—Ä—å —Å –º–æ–¥–µ–ª—å llama-cpp-python
        system_prompt: str,  # —Å–∏—Å—Ç–µ–º–Ω—ã–π –ø—Ä–æ–º—Ç
        support_system_role: bool,  # –ø–æ–¥–¥–µ—Ä–∂–∏—ã–≤–∞–µ—Ç –ª–∏ –º–æ–¥–µ–ª—å —Å–∏—Å—Ç–µ–º–Ω—ã–π –ø—Ä–æ–º—Ç
        history_len: int,  # –∫–æ–ª-–≤–æ –ø—Ä–µ–¥—ã–¥—É—â–∏—Ö —Å–æ–æ–±—â–µ–Ω–∏–π –∫–æ—Ç–æ—Ä—ã–µ —É—á–∏—Ç—ã–≤–∞—é—Ç—Å—è –≤ –∏—Å—Ç–æ—Ä–∏–∏
        do_sample: bool,  # –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –ª–∏ —Å–ª—É—á–∞–π–Ω–æ–µ —Å–µ–º–ø–ª–∏—Ä–æ–≤–∞–Ω–∏–µ –ø—Ä–∏ –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ –æ—Ç–≤–µ—Ç–∞ –º–æ–¥–µ–ª—å—é
        *generate_args,  # –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ –æ—Ç–≤–µ—Ç–∞ –º–æ–¥–µ–ª–∏
        ):

    # –ø–æ–ª—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏
    model = model_dict.get('model')
    # –ø–æ–ª—É—á–µ–Ω–∏–µ —Å–æ–æ–±—â–µ–Ω–∏—è –∏–∑ —á–∞—Ç –±–æ—Ç–∞
    user_message = chatbot[-1][0]
    # —Å–ø–∏—Å–æ–∫ —Å –ø–µ—Ä–µ–ø–∏—Å–∫–∞–º–∏ —é–∑–µ—Ä–∞ –∏ –±–æ—Ç–∞
    messages = []

    # –æ–±–Ω–æ–≤–ª–µ–Ω–∏–µ —Å–ª–æ–≤–∞—Ä—è –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ —Ç–µ–∫—Å—Ç–∞
    gen_kwargs = dict(zip(GENERATE_KWARGS.keys(), generate_args))
    gen_kwargs['top_k'] = int(gen_kwargs['top_k'])

    # –æ—Ç–∫–ª—é—á–µ–Ω–∏–µ —Å–ª—É—á–∞–π–Ω–æ–≥–æ —Å–µ–º–ø–ª–∏—Ä–æ–≤–∞–Ω–∏—è –µ—Å–ª–∏ do_sample=False
    if not do_sample:
        gen_kwargs['top_p'] = 0.0
        gen_kwargs['top_k'] = 1
        gen_kwargs['repeat_penalty'] = 1.0

    # –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ —Å–∏—Å—Ç–µ–º–Ω–æ–≥–æ –ø—Ä–æ–º—Ç–∞ –µ—Å–ª–∏ –æ–Ω –µ—Å—Ç—å –∏ –ø–æ–¥–¥–µ—Ä–∂–∏–≤–∞–µ—Ç—Å—è –º–æ–¥–µ–ª—å—é
    if support_system_role and system_prompt:
        messages.append({'role': 'system', 'content': system_prompt})

    # –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ –∏—Å—Ç–æ—Ä–∏–∏ –ø–µ—Ä–µ–ø–∏—Å–∫–∏ –≤ –ø—Ä–æ–º—Ç
    if history_len != 0:
        for user_msg, bot_msg in chatbot[:-1][-history_len:]:
            print(user_msg, bot_msg)
            messages.append({'role': 'user', 'content': user_msg})
            messages.append({'role': 'assistant', 'content': bot_msg})

    # –¥–æ–±–∞–≤–ª–µ–Ω–∏—è —Ç–µ–∫—É—â–µ–≥–æ —Å–æ–æ–±—â–µ–Ω–∏—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
    messages.append({'role': 'user', 'content': user_message})

    # —Ñ–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–∏–µ –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä–∞ –¥–ª—è –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ –æ—Ç–≤–µ—Ç–∞ –º–æ–¥–µ–ª—å—é —Å —Ñ–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–∏–µ–º –ø—Ä–æ–º—Ç–∞
    stream_response = model.create_chat_completion(
        messages=messages,
        stream=True,
        **gen_kwargs,
        )

    # –ø—É—Å—Ç—É—é —Å—Ç—Ä–æ–∫—É –±—É–¥–µ–º –∫–æ–Ω–∫–∞—Ç–µ–Ω–∏—Ä–æ–≤–∞—Ç—å —Å —Ç–æ–∫–µ–Ω–∞–º–∏ –æ—Ç–≤–µ—Ç–∞ –º–æ–¥–µ–ª–∏
    chatbot[-1][1] = ''
    # –∏—Ç–µ—Ä–∞—Ü–∏—è –∏ –≥–µ–Ω–µ—Ä–∞—Ü–∏—è —Ç–æ–∫–µ–Ω–æ–≤ –æ—Ç–≤–µ—Ç–∞ –º–æ–¥–µ–ª–∏ –≤ —Ü–∏–∫–ª–µ
    for chunk in stream_response:
        token = chunk['choices'][0]['delta'].get('content')
        if token is not None:
            chatbot[-1][1] += token
            yield chatbot


# –ø–æ–ª—É—á–µ–Ω–∏–µ –æ–∫–æ—à–∫–∞ –¥–ª—è —Å–∏—Å—Ç–µ–º–Ω–æ–≥–æ –ø—Ä–æ–º—Ç–∞. –µ—Å–ª–∏ interactive=True —Ç–æ –º–æ–∂–Ω–æ –≤–≤–æ–¥–∏—Ç—å —Å–∏—Å—Ç–µ–º–Ω—ã–π –ø—Ä–æ–º—Ç
# –ø—Ä–∏–º–µ—Ä—ã —Å–∏—Å—Ç–µ–º–Ω—ã—Ö –ø—Ä–æ–º—Ç–æ–≤:
# –û—Ç–≤–µ—á–∞–π –Ω–∞ –≤–æ–ø—Ä–æ—Å—ã –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –∫—Ä–∞—Ç–∫–æ, –µ—Å–ª–∏ –Ω–µ –∑–Ω–∞–µ—à—å –æ—Ç–≤–µ—Ç - –≥–æ–≤–æ—Ä–∏ "–Ω–µ –∑–Ω–∞—é".
# –ï—Å–ª–∏ –≤ –≤–æ–ø—Ä–æ—Å–µ –µ—Å—Ç—å —Ö–æ—Ç—å –º–∞–ª–µ–π—à–∏–π –Ω–∞–º–µ–∫ –Ω–∞ –ø–æ–ª–∏—Ç–∏–∫—É, –Ω–∞—Å–∏–ª–∏–µ, —Ä—É–≥–∞—Ç–µ–ª—å—Å—Ç–≤–∞, –æ—Ç–≤–µ—Ç—å —Ç–∞–∫: "–ù–∞ —ç—Ç—É —Ç–µ–º—É —è –Ω–µ –º–æ–≥—É –≥–æ–≤–æ—Ä–∏—Ç—å".
def get_system_prompt_component(interactive: bool) -> gr.Textbox:
    value = '' if interactive else 'System prompt is not supported by this model'
    return gr.Textbox(value=value, label='System prompt', interactive=interactive)


# –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—ã –Ω–∞—Å—Ç—Ä–æ–π–∫–∏ –∫–æ–Ω—Ñ–∏–≥–∞ –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ —Ç–µ–∫—Å—Ç–∞
def get_generate_args(do_sample: bool) -> List[gr.component]:
    # –µ—Å–ª–∏ do_sample –≤–∫–ª—é—á–µ–Ω (—ç–ª–µ–º–µ–Ω—Ç gr.Checkbox() –∞–∫—Ç–∏–≤–µ–Ω) —Ç–æ –æ—Ç–æ–±—Ä–∞–∂–∞—Ç—å —Å–ª–∞–π–¥–µ—Ä–∞ —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏ –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏
    visible = do_sample
    generate_args = [
        gr.Slider(label='temperature', value=GENERATE_KWARGS['temperature'], minimum=0.1, maximum=3, step=0.1, visible=visible),
        gr.Slider(label='top_p', value=GENERATE_KWARGS['top_p'], minimum=0.1, maximum=1, step=0.1, visible=visible),
        gr.Slider(label='top_k', value=GENERATE_KWARGS['top_k'], minimum=1, maximum=50, step=5, visible=visible),
        gr.Slider(label='repeat_penalty', value=GENERATE_KWARGS['repeat_penalty'], minimum=1, maximum=5, step=0.1, visible=visible),
    ]
    return generate_args


# ======================== –ü–ï–†–ï–ú–ï–ù–ù–´–ï ==============================

# –ø—É—Ç–∏ –∫—É–¥–∞ –±—É–¥—É—Ç —Å–∫–∞—á–∏–≤–∞—Ç—å—Å—è –º–æ–¥–µ–ª–∏
MODELS_PATH = Path('models')
MODELS_PATH.mkdir(exist_ok=True)

# —Å—Å—ã–ª–∫–∞ –Ω–∞ –º–æ–¥–µ–ª—å GGUF –¥–ª—è —Å—Ç–∞—Ä—Ç–æ–≤–æ–π –º–æ–¥–µ–ª–∏
DEFAULT_GGUF_URL = 'https://huggingface.co/bartowski/gemma-2-2b-it-GGUF/resolve/main/gemma-2-2b-it-Q8_0.gguf'

# –∑–∞–≥—Ä—É–∑–∫–∞ –∏ –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è —Å—Ç–∞—Ä—Ç–æ–≤–æ–π –º–æ–¥–µ–ª–∏, —Ñ–ª–∞–≥–∞ –ø–æ–¥–¥–µ—Ä–∂–∏–∫–∏ —Å–∏—Å—Ç–µ–º–Ω–æ–≥–æ –ø—Ä–æ–º—Ç–∞ –∏ –ª–æ–≥–æ–≤ –∑–∞–≥—Ä—É–∑–∫–∏
start_model_dict, start_support_system_role, start_load_log = download_gguf_and_init_model(
    gguf_url=DEFAULT_GGUF_URL, model_dict={},
    )

# —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –∫–æ–Ω—Ñ–∏–≥–∞ –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ —Ç–µ–∫—Å—Ç–∞
GENERATE_KWARGS = dict(
    temperature=0.2,  # —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞ –¥–ª—è —Å–æ—Ñ—Ç–º–∞–∫—Å–∞
    top_p=0.95,  # —Å—É–º–º–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π —Ç–æ–∫–µ–Ω–æ–≤ –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö –Ω—É–∂–Ω–æ –≤—ã–±–∏—Ä–∞—Ç—å —Å–ª–µ–¥—É—é—â–∏–π —Ç–æ–∫–µ–Ω
    top_k=40,  # –∏–∑ —Å–∫–æ–ª—å–∫–∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ –≤–µ—Ä–æ—è—Ç–Ω—ã—Ö —Ç–æ–∫–µ–Ω–æ–≤ –≤—ã–±–∏—Ä–∞—Ç—å —Å–ª–µ–¥—É—é—â–∏–π —Ç–æ–∫–µ–Ω
    repeat_penalty=1.0,  # —à—Ç—Ä–∞—Ñ –º–æ–¥–µ–ª–∏ –∑–∞ –ø–æ–≤—Ç–æ—Ä–µ–Ω–∏—è
    )

# ====================== –ò–ù–¢–ï–†–§–ï–ô–° –ü–†–ò–õ–û–ñ–ï–ù–ò–Ø ============================

# —Ç–µ–º–∞ –æ—Ñ–æ—Ä–º–ª–µ–Ω–∏—è –∏ CSS
# theme = gr.themes.Monochrome()
theme = gr.themes.Base(primary_hue='green', secondary_hue='yellow', neutral_hue='zinc').set(
    loader_color='rgb(0, 255, 0)',
    slider_color='rgb(0, 200, 0)',
    body_text_color_dark='rgb(0, 200, 0)',
    button_secondary_background_fill_dark='green',
)
css = '''.gradio-container {width: 60% !important}'''

# –Ω–∞—á–∞–ª–æ –æ–ø–∏—Å–∞–Ω–∏—è –∏–Ω—Ç–µ—Ä—Ñ–µ–π—Å–∞ –ø—Ä–∏–ª–æ–∂–µ–Ω–∏—è
with gr.Blocks(theme=theme, css=css) as interface:

    # —Å–ª–æ–≤–∞—Ä—å —Å —Ç–µ–∫—É—â–µ–π –º–æ–¥–µ–ª—å—é llama-cpp-python
    model_dict = gr.State(start_model_dict)
    # —Ñ–ª–∞–≥ –ø–æ–¥–¥–µ—Ä–∂–∫–∏ —Å–∏—Å—Ç–µ–º–Ω–æ–≥–æ –ø—Ä–æ–º—Ç–∞ –º–æ–¥–µ–ª—å—é
    support_system_role = gr.State(start_support_system_role)

    # ========== –°–¢–†–ê–ù–ò–¶–ê –ß–ê–¢-–ë–û–¢–ê ===============================

    with gr.Tab('Chat bot'):
        with gr.Row():
            with gr.Column(scale=3):
                # –æ–∫–æ—à–∫–æ —á–∞—Ç-–±–æ—Ç–∞ –∏ –æ–∫–æ—à–∫–æ –≤–≤–æ–¥–∞ —Å–æ–æ–±—â–µ–Ω–∏—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
                chatbot = gr.Chatbot(show_copy_button=True, bubble_full_width=False, height=480)
                user_message = gr.Textbox(label='User')

                # –∫–Ω–æ–ø–∫–∏ –æ—Ç–ø—Ä–∞–≤–∏—Ç—å —Å–æ–æ–±—â–µ–Ω–∏–µ, —Å—Ç–æ–ø –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ –∏ —É–¥–∞–ª–∏—Ç—å –∏—Å—Ç–æ—Ä–∏—é —á–∞—Ç–∞
                with gr.Row():
                    user_message_btn = gr.Button('Send')
                    stop_btn = gr.Button('Stop')
                    clear_btn = gr.Button('Clear')

                # –æ–∫–æ—à–∫–æ –¥–ª—è —Å–∏—Å—Ç–µ–º–Ω–æ–≥–æ –ø—Ä–æ–º—Ç–∞
                system_prompt = get_system_prompt_component(interactive=support_system_role.value)

            with gr.Column(scale=1, min_width=80):
                with gr.Group():
                    # –¥–ª–∏–Ω–∞ –∏—Å—Ç–æ—Ä–∏–∏ –∫–æ—Ç–æ—Ä—É—é –±—É–¥–µ—Ç —É—á–∏—Ç—ã–≤–∞—Ç—å –º–æ–¥–µ–ª—å
                    gr.Markdown('Length of message history')
                    history_len = gr.Slider(
                        minimum=0,
                        maximum=10,
                        value=0,
                        step=1,
                        info='Number of previous messages taken into account in history',
                        label='history_len',
                        show_label=False,
                        )

                    with gr.Group():
                        gr.Markdown('Generation parameters')
                        # –ø–µ—Ä–µ–∫–ª—é—á–∞—Ç–µ–ª—å –∞–∫—Ç–∏–≤–∞—Ü–∏–∏ —Å–ª—É—á–∞–π–Ω–æ–≥–æ —Å–µ–º–ø–ª–∏—Ä–æ–≤–∞–Ω–∏—è –ø—Ä–∏ –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ —Ç–µ–∫—Å—Ç–∞ –º–æ–¥–µ–ª—å—é
                        do_sample = gr.Checkbox(
                            value=False,
                            label='do_sample',
                            info='Activate random sampling',
                            )
                        # –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ —Ç–µ–∫—Å—Ç–∞
                        generate_args = get_generate_args(do_sample.value)
                        do_sample.change(
                            fn=get_generate_args,
                            inputs=do_sample,
                            outputs=generate_args,
                            show_progress=False,
                        )

        # ------------------ –ö–ù–û–ü–ö–ò –û–¢–ü–†–ê–í–ò–¢–¨ –û–ß–ò–°–¢–ò–¢–¨ –ò –°–¢–û–ü ------------

        # –Ω–∞–∂–∞—Ç–∏–µ Enter –∏ –∫–Ω–æ–ø–∫–∞ –æ—Ç–ø—Ä–∞–≤–∏—Ç—å
        generate_event = gr.on(
            triggers=[user_message.submit, user_message_btn.click],
            fn=user_message_to_chatbot,
            inputs=[user_message, chatbot],
            outputs=[user_message, chatbot],
        ).then(
            fn=bot_response_to_chatbot,
            inputs=[chatbot, model_dict, system_prompt, support_system_role, history_len, do_sample, *generate_args],
            outputs=[chatbot],
        )
        # –∫–Ω–æ–ø–∫–∞ –°—Ç–æ–ø
        stop_btn.click(
            fn=None,
            inputs=None,
            outputs=None,
            cancels=generate_event,
        )
        # –∫–Ω–æ–ø–∫–∞ –û—á–∏—Å—Ç–∏—Ç—å —á–∞—Ç
        clear_btn.click(
            fn=lambda: None,
            inputs=None,
            outputs=[chatbot],
            )

    # ============= –°–¢–†–ê–ù–ò–¶–ê –ó–ê–ì–†–£–ó–ö–ò –ú–û–î–ï–õ–ï–ô ============================

    with gr.Tab('Load model'):
        # –æ–∫–Ω–æ –¥–ª—è URL —Å—Å—ã–ª–∫–∏ –Ω–∞ –º–æ–¥–µ–ª—å –≤ —Ñ–æ—Ä–º–∞—Ç–µ GGUF
        gguf_url = gr.Textbox(
            value='',
            label='Link to GGUF',
            placeholder='URL link to the model in GGUF format',
            )
        # –∫–Ω–æ–ø–∫–∞ –∑–∞–≥—Ä—É–∑–∫–∏ –º–æ–¥–µ–ª–∏ –ø–æ URL —Å—Å—ã–ª–∫–µ
        load_model_btn = gr.Button('Downloading GGUF and initializing the model')

        # —Å—Ç–∞—Ç—É—Å –∑–∞–≥—Ä—É–∑–∫–∏ –∏ –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏–∏ –º–æ–¥–µ–ª–∏
        load_log = gr.Textbox(
            value=start_load_log,
            label='Model loading status',
            lines=3,
            )

        # –Ω–∞–∂–∞—Ç–∏–µ –∫–Ω–æ–ø–∫–∏ –∑–∞–≥—Ä—É–∑–∫–∏ –∏ –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏–∏ –º–æ–¥–µ–ª–∏
        load_model_btn.click(
            fn=download_gguf_and_init_model,
            inputs=[gguf_url, model_dict],
            outputs=[model_dict, support_system_role, load_log],
        ).success(
            fn=get_system_prompt_component,
            inputs=[support_system_role],
            outputs=[system_prompt],
        )

# –∑–∞–ø—É—Å–∫ –ø—Ä–∏–ª–æ–∂–µ–Ω–∏—è –≤ Colab
interface.launch(debug=True)

# –∑–∞–ø—É—Å–∫ –ø—Ä–∏–ª–æ–∂–µ–Ω–∏—è –ø—Ä–∏ –¥–µ–ø–ª–æ–µ
# interface.launch(server_name='0.0.0.0', server_port=7860)
````
